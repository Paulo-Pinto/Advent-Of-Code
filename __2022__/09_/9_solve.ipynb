```python
import re
import numpy as np

def tail_catch_up(h, t):
    dist = abs(h.real-t.real) - abs(h.imag-t.imag)
    print(f"distance btw {h=} and {t=} , {dist}")

def is_diagonal(h,t):
    return h.imag != t.imag and h.real != t.real

def read_file(filename="test"):    
    # head = 11+16j
    visited = [10+16j]
    ropes = [10+16j for _ in range(10)]
    # ropes_moved = []
    
    with open(filename) as fp:
        for line in fp.read().splitlines():
            direction, power = re.split(" ", line)
            print(direction,power)

            for _ in range(0, int(power)):                
                # first element is head, last is tail / 9
                for idx, rope in enumerate(ropes):
                    # move head
                    if idx == 0:
                        match(direction):
                            case "R": ropes[idx] += 1+0j
                            case "L": ropes[idx] -= 1+0j
                            case "D": ropes[idx] += 0+1j
                            case "U": ropes[idx] -= 0+1j
                        continue
                    
                    # move rest
                    h = ropes[idx - 1]
                    t = ropes[idx]
                    # print(f"before {t=}")
                    # print(h, t)
                    if abs(t.real - h.real) > 1 or abs(t.imag - h.imag) > 1:
                        match(direction):
                            case "R":
                                if is_diagonal(h, t):
                                    t += 0 + (h.imag - t.imag)/abs((h.imag - t.imag)) * 1j                        
                                t += 1+0j

                            case "L":
                                if is_diagonal(h, t):
                                    t += 0 + (h.imag - t.imag)/abs((h.imag - t.imag)) * 1j                        
                                t -= 1+0j
 
                            case "D":                             
                                if is_diagonal(h, t):
                                    t += (h.real - t.real)/abs(h.real - t.real) + 0j                           
                                t += 0+1j

                            case "U":                            
                                if is_diagonal(h, t):
                                    t += (h.real - t.real)/abs((h.real - t.real)) + 0j                           
                                t -= 0+1j
                    # print(f"after {t=}")
                    ropes[idx] = t
                    if h == 0+4j:
                        break

                    visited.append(ropes[-1])
                    visited = list(set(visited))
                complex_to_grid(visited)

                # print(ropes)
                
            # print(f"{h=} {t=}")

        # print(f"{h=} {t=}")
        
        print(len(set(visited)))


def complex_to_grid(l):
    wide, tall = 28, 23
    grid = [["."]*wide for _ in range(tall)]
    for r in range(tall):
        for c in range(wide):
            grid[r][c] == "."

    for idx, n in enumerate(l):
        
        i, r = int(n.imag), int(n.real)
        if idx == 0:
            grid[i][r] = "H"
            continue
        grid[i][r] = "#"
    print("\n\n")
    print(np.matrix(grid))

np.set_printoptions(edgeitems=25, linewidth=100000)
read_file("test")
```
