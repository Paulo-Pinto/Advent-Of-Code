```python
import re
from collections import defaultdict
class Tree:
  def __init__(self, data):
    self.data = data
    self.children = []

tree = Tree("/")

def find_node(data:str):
    if tree.data == data: return tree
     
    for c in tree.children:
        return find_node(c)
```

```python

children_of = {} # dependencies
size_of = defaultdict(int)  #

def recursive(parent):
    # print(f"recursiva {parent}")
    
    if parent not in children_of.keys():
        return 0

    t = 0
    for c in children_of[parent]:
        # print(size_of[c])
        t+= size_of[c] + recursive(c)
    return t
    
def read_file(filename="test"):
    curr_dir = "/"

    with open(filename) as fp:
        for line in fp.read().splitlines():

            if (num := re.findall(r"\d+", line)):
                size_of[curr_dir] += int(num[0])

            if "$" in line:  # command
                if "cd" in line:
                    curr_dir = line.split("$ cd ")[-1]
                    children_of[curr_dir] = []

            if "dir" in line:
                children_of[curr_dir].append(line.split("dir ")[-1])

        # print(f"{size_of=}")
        # print(f"{children_of=}")

        t = 0
        for child in children_of.keys():
            size_of[child] += recursive(child)
            print(size_of[child])
            if size_of[child] < 100000:
                t += size_of[child]

        return t
read_file("input")
```
