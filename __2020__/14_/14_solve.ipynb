```python
from collections import defaultdict

def read_file(type = "test"):

    iters = {}
    mems = []
    mask = [0]*36
    ctr = 0

    with open(type) as fp:
        while (line := fp.readline()):
            match(line[:4]):
                case "mask":
                    if ctr > 0:
                        iters[ctr] = (mask, mems)
                    _, mask = line.strip("\n").split(" = ")
                    ctr += 1

                case "mem[":
                    mem, mem_value = line.strip("\n").split(" = ")
                    mems.append((mem[4], mem_value))
        
        iters[ctr] = (mask, mems)

    return iters

def address_to_bin(v):
    return int("".join([str(e) for e in v]), 2)

print(read_file("input"))
```

```python
for mask, mems in read_file("input").values():

    print(mask, mems)
    addresses = defaultdict(lambda: [0] * 36)

    for address, value in mems:
        print(address, value)

        binary = bin(int(value))[2:]
        binary = f"{binary:0>36}"

        # print(binary)
        # print(binary)
        # print(mask)

        for i in range(len(binary)):
            # print(binary[i], mask[i])
            if binary[i] == mask[i] or mask[i] == "X":
                addresses[address][i] = int(binary[i])
                # print("okay")
                # print(addresses[address])
            else:
                addresses[address][i] = int(mask[i])
                # print(address)
                # print("changed")
                # print(binary, mask)

    soma = 0
    for v in addresses.values():
        print(v)
        soma += address_to_bin(v)
    print(soma)
```
