```python
def read_file(filename):
    with open(filename) as fp:
        
        data = []
        for line in fp.readlines():

            # if len(line) == 1: # blank line / new item
            #     continue
                
            cmd, score = line.strip().split(" ")
            data.append((cmd, int(score)))
            
        return data
```

```python
data = read_file("input")

def find(accu = 0, pos = 0, altered = 0, new_cmd=""):
    # try replacing new location
    if altered:
        data[pos] = new_cmd

    ctr = 0
    while 1:
        ctr += 1

        if ctr > 1000:
            break

        if pos == len(data):
            print(f"found it, {accu=}")
            return 1

        cmd, v = data[pos]

        match(cmd):
            case "nop":
                if not altered:
                    if find(accu, pos, altered = 1, new_cmd = ("jmp", v)):
                        break
                pass
            case "acc":
                accu += v
            case "jmp":
                # print(pos, pos+v)
                if not altered:
                    if find(accu, pos, seen, altered = 1, new_cmd = ("nop", v)):
                        break
                pos += (v - 1)
        
        pos += 1

find()
```
