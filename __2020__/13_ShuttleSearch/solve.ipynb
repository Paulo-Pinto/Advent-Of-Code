## Part one

```python
def read_file(type = "simple"):
    with open(f"input_{type}") as fp:
        earliest = int(fp.readline().strip("\n"))
        bus_id = [int(x) for x in fp.readline().strip("\n").split(",") if x != "x"]
        return earliest, bus_id
```

```python
earliest, bus_id = read_file("normal")
curr_time = earliest
answer = None

while answer == None:
    for i in bus_id:
        if curr_time % i == 0:
            answer = curr_time, i
            break

    curr_time += 1

print(curr_time - earliest - 1, i)
print(f"answer {(curr_time - earliest - 1) * i}")
```

```python
# doesn't return first line
def read_file_pt2(type = "simple"):
    with open(f"input_{type}") as fp:
        fp.readline()
        return [x for x in fp.readline().strip("\n").split(",")]
```

```python
def check_mod_new(curr_timestamp, val, relative_idx):
    return (curr_timestamp + relative_idx) % val == 0

l = read_file_pt2("normal")

max_ele = max([int(m) for m in l if m.isdigit()])
idx = l.index(str(max_ele))
# start looking at the max_element minus its index
timestamp = max_ele

new_list = [(max_ele, 0)]

ctr = 0

# biggest to end
for x in range(idx + 1, len(l)):
    ctr += 1
    if l[x] != "x":
        new_list.append((int(l[x]), ctr))

ctr = 0

for x in range(idx - 1, -1, -1):
    ctr += 1
    if l[x] != "x":
        new_list.insert(0, (int(l[x]), -ctr))

print(*new_list)
print(f"Starting at {timestamp}")
print(f"{max_ele=} @ {idx=}")

# 1000000000000000
goal = 1000000000000000
print(goal)

while timestamp < goal * 2:

    timestamp += max_ele # increment steps by max_element

    if (timestamp) % (max_ele * 1_000_000) == 0:
        print(timestamp)
 
    for v,ri in new_list: # val, relative index
        if not check_mod_new(timestamp, v, ri): break
    else:
        print(f"final {timestamp}")
        timestamp = 1000000000000001
```

```python
def read_file_pt2_x_as_1(type = "simple"):
    with open(f"input_{type}") as fp:
        fp.readline()
        return [1 if x == "x" else int(x) for x in fp.readline().strip("\n").split(",")]
```

```python
bus_ids = read_file_pt2_x_as_1("normal")
print(bus_ids)
timestamp = 0
ctr = 0
step = bus_ids[ctr]

for i in range(1, len(bus_ids)):
    print(f"seeking {bus_ids[i]}")

    while (timestamp + i) % bus_ids[i] != 0:
        timestamp += step

    step *= bus_ids[i]

    print(timestamp, step)    
```
