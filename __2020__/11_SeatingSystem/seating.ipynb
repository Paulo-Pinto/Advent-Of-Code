```python
import numpy as np
from copy import deepcopy

def read_file(type="normal"):
    with open(f"input_{type}") as fp:
        return [list(x.strip("\n")) for x in fp.readlines()]

map = read_file()
```

# Part one

```python
def occupied_adjacent(row, col, radius=1):
    rows, cols = len(map), len(map[0])
    neighbours = []

    for i in range(row - radius, row + radius + 1):
        for j in range(col - radius, col + radius + 1):

            # ignore center element
            if i == row and j == col:
                continue

            if 0 <= i < rows and 0 <= j < cols:
                neighbours.append(map[i][j])

    return neighbours.count("#")

def occupy_seats():
    global map, ctr
    new_map = deepcopy(map)

    ctr += 1

    for row in range(len(map)):
        for col in range(len(map[0])):

            if map[row][col] == ".":
                continue

            # print(row, col)
            if map[row][col] == "L":
                if occupied_adjacent(row, col) == 0:
                    # occupy seat
                    new_map[row][col] = "#"
                    continue

            if map[row][col] == "#":
                if occupied_adjacent(row, col) >= 4:
                    # leave seat
                    new_map[row][col] = "L"
                    continue

    if map == new_map:
        print(f"stopped at {ctr} iterations")
        return 0

    map = new_map
    # print(np.matrix(map))
    return 1

def count_occupied():
    return sum([row.count("#") for row in map]) # count occupied
```

```python
map = read_file()
ctr = 0

while occupy_seats():
    pass
else:
    print(f"{count_occupied()} occupied seats")
```

# Part two

```python
# checks next seat in 8 different directions, there's probably an easier way to do this
def occupied_propagate(row, col, verbose = False):
    neighbours = []
    spots = {
        "up" : 0,
        "down" : 0,
        "left" : 0,
        "right" : 0,
        "upleft" : 0,
        "upright" : 0,
        "downleft" : 0,
        "downright" : 0,
    }

    # horizontal
    x = 1
    while x <= len(map[row]) - 1:
        l, r = col - x, col + x

        if l >= 0 and spots["left"] == 0:
            if map[row][l] != ".":
                if verbose: print(f"seat at distance {x} LEFT")
                neighbours.append(map[row][l])
                spots["left"] = 1

        if r < len(map[row]) and spots["right"] == 0:
            if map[row][r] != ".":
                if verbose: print(f"seat at distance {x} RIGHT")
                neighbours.append(map[row][r])
                spots["right"] = 1

        x += 1
    
    # vertical
    y = 1
    while y <= len(map) - 1:
        u, d = row - y, row + y

        if u >= 0 and spots["up"] == 0:
            if map[u][col] != ".":
                if verbose: print(f"seat at distance {y} UP")
                neighbours.append(map[u][col])
                spots["up"] = 1

        if d < len(map) and spots["down"] == 0:
            if map[d][col] != ".":
                if verbose: print(f"seat at distance {y} DOWN")
                neighbours.append(map[d][col])
                spots["down"] = 1

        y += 1

    # diagonal ↖ ↘
    y = 1
    x = 1
    while y <= len(map[row]) - 1 and x <= len(map) - 1:
        u, d = row - y, row + y
        l, r = col - x, col + x

        if u >= 0 and l >= 0 and spots["upleft"] == 0:
            if map[u][l] != ".":
                if verbose: print(f"seat at distance {y} ↖")
                neighbours.append(map[u][l])
                spots["upleft"] = 1

        if d < len(map) and r < len(map[row]) and spots["downright"] == 0:
            if map[d][r] != ".":
                if verbose: print(f"seat at distance {y} ↘")
                neighbours.append(map[d][r])
                spots["downright"] = 1

        y += 1
        x += 1

    # diagonal ↗ ↙
    y = 1
    x = 1
    while y <= len(map[row]) - 1 and x <= len(map) - 1:
        u, d = row - y, row + y
        l, r = col + x, col - x # mirrow x axis

        if u >= 0 and l < len(map[row]) and spots["upright"] == 0:
            if map[u][l] != ".":
                if verbose: print(f"seat at distance {y} ↗")
                neighbours.append(map[u][l])
                spots["upright"] = 1

        if d < len(map) and r >= 0 and spots["downleft"] == 0:
            if map[d][r] != ".":
                if verbose: print(f"seat at distance {y} ↙")
                neighbours.append(map[d][r])
                spots["downleft"] = 1

        y += 1
        x += 1

    # print(neighbours)

    return neighbours.count("#")

map = read_file("tiny")
print(np.matrix(map))
occupied_propagate(4, 3, True)
```

```python
def occupy_seats_pt2():
    global map, ctr
    new_map = deepcopy(map)
    # print(np.matrix(map))

    ctr += 1
    print(ctr)
    for row in range(len(map)):
        for col in range(len(map[0])):

            if map[row][col] == ".":
                continue

            # print(row, col)
            if map[row][col] == "L":
                abou = occupied_propagate(row, col)
                # print(f"propagating L {abou}")
                if abou == 0:
                    # occupy seat
                    new_map[row][col] = "#"
                    continue

            if map[row][col] == "#":
                abou = occupied_propagate(row, col)
                # print(f"propagating # {abou}")
                if abou >= 5:
                    # leave seat
                    new_map[row][col] = "L"
                    continue

    if map == new_map:
        print(f"stopped at {ctr} iterations")
        return 0

    map = new_map
    # input("")
    print(f"{count_occupied()} occupied seats")

    return 1
```

```python
map = read_file("normal")
ctr = 0

while occupy_seats_pt2() and ctr < 3000:
    pass
else:
    print(f"{count_occupied()} occupied seats")
```
