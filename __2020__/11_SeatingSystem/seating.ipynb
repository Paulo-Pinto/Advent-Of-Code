```python
import numpy as np
from copy import deepcopy

def read_file(type="normal"):
    with open(f"input_{type}") as fp:
        return [list(x.strip("\n")) for x in fp.readlines()]

map = read_file()
```

# Part one

```python
def occupied_adjacent(row, col, radius=1):
    rows, cols = len(map), len(map[0])
    neighbours = []

    for i in range(row - radius, row + radius + 1):
        for j in range(col - radius, col + radius + 1):

            # ignore center element
            if i == row and j == col:
                continue

            if 0 <= i < rows and 0 <= j < cols:
                neighbours.append(map[i][j])

    return neighbours.count("#")

def occupy_seats():
    global map, ctr
    new_map = deepcopy(map)

    ctr += 1

    for row in range(len(map)):
        for col in range(len(map[0])):

            if map[row][col] == ".":
                continue

            # print(row, col)
            if map[row][col] == "L":
                if occupied_adjacent(row, col) == 0:
                    # occupy seat
                    new_map[row][col] = "#"
                    continue

            if map[row][col] == "#":
                if occupied_adjacent(row, col) >= 4:
                    # leave seat
                    new_map[row][col] = "L"
                    continue

    if map == new_map:
        print(f"stopped at {ctr} iterations")
        return 0

    map = new_map
    # print(np.matrix(map))
    return 1

def count_occupied():
    return sum([row.count("#") for row in map]) # count occupied
```

```python
map = read_file()
ctr = 0

while occupy_seats():
    pass
else:
    print(f"{count_occupied()} occupied seats")
```

# Part two

```python
def occupied_propagate(row, col, verbose = False):
    neighbours = []

    # horizontal
    x = 1
    while x <= len(map[row]) - 1:
        l, r = col - x, col + x

        if l >= 0:
            if map[row][l] == "#":
                if verbose: print(f"ocupado a {x} lugares à ESQ")
                neighbours.append(map[row][l])

        if r < len(map[row]):
            if map[row][r] == "#":
                if verbose: print(f"ocupado a {x} lugares à DIR")
                neighbours.append(map[row][r])
        x += 1
    
    # vertical
    y = 1
    while y <= len(map) - 1:
        u, d = row - y, row + y

        if u >= 0:
            if map[u][col] == "#":
                if verbose: print(f"ocupado a {y} lugares UP")
                neighbours.append(map[u][col])

        if d < len(map[row]):
            if map[d][col] == "#":
                if verbose: print(f"ocupado a {y} lugares DOWN")
                neighbours.append(map[d][col])

        y += 1

    # diagonal ↖ ↘
    y = 1
    x = 1
    while y <= len(map[row]) - 1 and x <= len(map) - 1:
        u, d = row - y, row + y
        l, r = col - x, col + x

        if u >= 0 and l >= 0:
            if map[u][l] == "#":
                if verbose: print(f"ocupado a {y} lugares ↖")
                neighbours.append(map[u][l])

        if d < len(map[row]) and r < len(map[row]):
            if map[d][r] == "#":
                if verbose: print(f"ocupado a {y} lugares ↘")
                neighbours.append(map[d][r])

        y += 1
        x += 1

    # diagonal ↗ ↙
    y = 1
    x = 1
    while y <= len(map[row]) - 1 and x <= len(map) - 1:
        u, d = row - y, row + y
        l, r = col + x, col - x # switched left and right to mirror

        if u >= 0 and l < len(map[row]):
            if map[u][l] == "#":
                if verbose: print(f"ocupado a {y} lugares ↗")
                neighbours.append(map[u][l])

        if d < len(map[row]) and r >= 0:
            if map[d][r] == "#":
                if verbose: print(f"ocupado a {y} lugares ↙")
                neighbours.append(map[d][r])

        y += 1
        x += 1

    return len(neighbours)

map = read_file("tiny")
print(np.matrix(map))
occupied_propagate(4, 3)
```

```python
def occupy_seats_pt2():
    global map, ctr
    new_map = deepcopy(map)

    ctr += 1
    print(ctr)
    for row in range(len(map)):
        for col in range(len(map[0])):

            if map[row][col] == ".":
                continue

            # print(row, col)
            if map[row][col] == "L":
                abou = occupied_propagate(row, col)
                # print(f"propagating L {abou}")
                if abou == 0:
                    # occupy seat
                    new_map[row][col] = "#"
                    continue

            if map[row][col] == "#":
                abou = occupied_propagate(row, col)
                # print(f"propagating # {abou}")
                if abou >= 5:
                    # print("propagated #")
                    # leave seat
                    new_map[row][col] = "L"
                    continue

    if map == new_map:
        print(f"stopped at {ctr} iterations")
        return 0

    map = new_map
    print(np.matrix(map))
    input("")
    return 1
```

```python
map = read_file("simple")
ctr = 0

while occupy_seats_pt2():
    pass
else:
    print(f"{count_occupied()} occupied seats")
```
