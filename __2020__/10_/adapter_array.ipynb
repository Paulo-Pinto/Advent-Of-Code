```python
def read_input(type="normal"):
    file = f"input_{type}.txt"

    with open(file) as fp:
        data = [int(x.strip("\n")) for x in fp.readlines()]
        data.append(0)  # add default 0
        data.sort()
        # data.append(data[len(data) - 1] + 3) # add last element (max + 3)
        data.append(max(data) + 3)  # add last element (max + 3)
        return data
```

## Part one

```python
def calc_distances(data):
    for pos, i in enumerate(data):
        for j in data[pos + 1 : pos + 4]:
            if (d := abs(i - j)) <= 3:
                yield d

            break


def part_one(data):
    distances = list(calc_distances(data))

    # calc distances
    ctr_dist_1 = distances.count(1)
    ctr_dist_3 = distances.count(3)
    print(f"{ctr_dist_1} * {ctr_dist_3} = {(ctr_dist_1 * ctr_dist_3)}")
    return ctr_dist_1 * ctr_dist_3


part_one(read_input("normal"))
```

## Part two

```python
# get combinations
data = read_input("normal")
distances = iter(calc_distances(data))

total = 1  # track total possibilities

# find distances of 1's in a row
while (ele := next(distances, -1)) != -1:

    if ele == 1:
        ctr = 1

        while next(distances) == 1:
            ctr += 1

        # increase possibilities count
        if ctr > 1:
            if ctr == 2:
                total *= 2
            if ctr > 2:
                total *= 3 * 2 ** (ctr - 3) + 1

print(total)
```
