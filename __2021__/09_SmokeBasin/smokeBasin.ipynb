```python
import numpy as np


def read_file(file):
    with open(file) as f:
        return [line.strip("\n") for line in f]


read_file("input_9_simple")
```

```python
def check_neighbours(data, line_length):
    lava_pit_ctr = 0
    for i in range(len(data)):
        curr = data[i]
        # print("\n\n" + curr)
        if i > line_length:
            # print("topleft, top, topright")
            # topleft = data[i - line_length - 1]
            # topright = data[i - line_length + 1]
            # print(topleft, top, topright)
            top = data[i - line_length]
            # print(f"top {top}")
            if top <= curr:
                # print("next")
                continue

        if i + line_length < len(data):
            # print("botleft, bot, botright")
            # botleft = data[i + line_length - 1]
            bot = data[i + line_length]
            # botright = data[i + line_length + 1]
            # print(botleft, bot, botright)
            # print(f"bot {bot}")
            if bot <= curr:
                # print("next")
                continue

        if i > 0:
            left = data[i - 1]
            # print(f"l {left}")
            if left <= curr:
                # print("next")
                continue

        if i + 1 < len(data):
            right = data[i + 1]
            # print(f"r {right}")
            if right <= curr:
                # print("next")
                continue
        print(i, curr)
        lava_pit_ctr += 1 + int(curr)
    return lava_pit_ctr
```

```python
d = read_file("input_9")
length = len(d[0])
# narr = np.array(aobu)
d = list("".join(d))

# for i in range(0,len(aobu),length):
#     print(aobu[i])
print(check_neighbours(d, length))
```
